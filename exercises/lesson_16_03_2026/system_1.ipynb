```python
import sys
sys.path.append("../../")

from lib.data.dataplot import *
from lib.utils.time import *

class S1:
    
    def __init__(self):
        self.y = 0
        
    def evaluate(self, delta_t, _input):
        y = self.y - 0.5 * delta_t * self.y + 3 * delta_t * _input
        self.y = y
        return y

dp = DataPlotter()
dp.set_x("time (seconds)")
dp.add_y("out", "Output")

s = S1()
u = 10

t = Time(use_fake_time=True)
t.start()
while t.get() < 10:

    t.sleep(0.001)
    delta_t = t.elapsed()
    
    y = s.evaluate(delta_t, u)
    
    u = 0
    
    dp.append_x(t.get())
    dp.append_y("out", y)

dp.plot()
```
