```python
import sys
sys.path.append("../../")

from lib.data.dataplot import *
from lib.utils.time import *

class S2:
    
    def __init__(self):
        self.x1 = 0
        self.x2 = 0
        
    def evaluate(self, delta_t, _input):
        new_x2 = self.x2 - delta_t * self.x2 - delta_t * self.x1 + delta_t * _input
        new_x1 = self.x1 + delta_t * self.x2
        self.x1 = new_x1
        self.x2 = new_x2
        return self.x1

dp = DataPlotter()
dp.set_x("time (seconds)")
dp.add_y("out", "Output")

s = S2()
u = 10

t = Time(use_fake_time=True)
t.start()
while t.get() < 30:

    t.sleep(0.001)
    delta_t = t.elapsed()
    
    y = s.evaluate(delta_t, u)
    u = 0
    
    dp.append_x(t.get())
    dp.append_y("out", y)

dp.plot()
```
