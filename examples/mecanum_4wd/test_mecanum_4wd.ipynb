```python
import sys
sys.path.append("../../")

#
# TO BE USED WITH godot/mecanum_4w
#

from lib.system.mecanum import *
from lib.dds.dds import *
from lib.utils.time import *

dds = DDS()
dds.start()

dds.subscribe(['tick'])

cart = MecanumFourWheels(10, # mass, 10kg 
                         1.0, # wheelbase
                         0.8, # lin-friction 
                         0.9, # ang-friction 
                         0.2) # 20cm radius motion wheel 

t1,t2,t3,t4 = 0.2, 0, 0, 0.2

t = Time()
t.start()
while t.get() < 3:

    dds.wait('tick')
    delta_t = t.elapsed()
    
    cart.evaluate(delta_t, t1, t2, t3, t4)
    pose = cart.get_pose()

    dds.publish('X', pose[0], DDS.DDS_TYPE_FLOAT)
    dds.publish('Y', pose[1], DDS.DDS_TYPE_FLOAT)
    dds.publish('Theta', pose[2], DDS.DDS_TYPE_FLOAT)

dds.stop()
```
