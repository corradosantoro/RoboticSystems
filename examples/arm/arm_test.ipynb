```python
import sys
sys.path.append("../../")

import time
import math

from lib.data.dataplot import *
from lib.system.arm import *
from lib.utils.time import *

dps = DataPlotter()
dps.set_x("time (seconds)")
dps.add_y("speed", "speed")

dpa = DataPlotter()
dpa.set_x("time (seconds)")
dpa.add_y("angle", "Angle")

arm = Arm(1.0, 0.5, 0.9)
# mass = 1 Kg
# length = 0.5 meters
# friction = 0.9

torque = 2.5 # Nm

t = Time()
t.start()
while t.get() < 20:

    time.sleep(0.01)
    delta_t = t.elapsed()
    (p, v) = arm.evaluate(delta_t, torque)

    dps.append_x(t.get())
    dps.append_y("speed", v)
    
    dpa.append_x(t.get())
    dpa.append_y("angle", p)

dpa.plot()
dps.plot()

```
