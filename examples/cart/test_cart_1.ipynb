```python
import sys
import time
sys.path.append("../../")

from lib.data.dataplot import *
from lib.utils.time import *
from lib.system.cart import *

spd = DataPlotter()
spd.set_x("time (seconds)")
spd.add_y("speed", "Speed")

pos = DataPlotter()
pos.set_x("time (seconds)")
pos.add_y("pos", "Position")

cart = Cart(1.0, 0.9)

force = 1000

t = Time()
t.start()
while t.get() < 5:
    
    time.sleep(0.01)
    delta_t = t.elapsed()
    
    (p, v) = cart.evaluate(delta_t, force)
    
    force = 0
    
    spd.append_x(t.get())
    spd.append_y("speed", v)
    
    pos.append_x(t.get())
    pos.append_y("pos", p)

spd.plot()
pos.plot()
```
