```python
import sys
sys.path.append("../../")

from lib.data.dataplot import *

import math

d = DataPlotter()
d.set_x("time (seconds)")
d.add_y("sin", "Sin(t)")
d.add_y("cos", "cos(t)")

t = 0
while t < 20:
    d.append_x(t)
    d.append_y("sin", math.sin(t))
    d.append_y("cos", math.cos(t))
    t += 0.01

d.plot()

```
